```python
orders = pd.read_csv('data/orders.csv', sep=';') #
order_details = pd.read_csv('data/order_details.csv', sep=';') #
products = pd.read_csv('data/products.csv', sep=';') #
categories = pd.read_csv('data/categories.csv', sep=';') #
customer_customer_demo = pd.read_csv('data/customer_customer_demo.csv', sep=';') # Not have data
customer_demographics = pd.read_csv('data/customer_demographics.csv', sep=';') # Not have data
customers = pd.read_csv('data/customers.csv', sep=';') #
employee_territories = pd.read_csv('data/employee_territories.csv', sep=';') #
employees = pd.read_csv('data/employees.csv', sep=';') #
region = pd.read_csv('data/region.csv', sep=';') #
shippers = pd.read_csv('data/shippers.csv', sep=';') # No have correlation
suppliers = pd.read_csv('data/suppliers.csv', sep=';') # 
territories = pd.read_csv('data/territories.csv', sep=';') #
us_states = pd.read_csv('data/us_states.csv', sep=';') # No have correlation
```

```python
df1 = orders.merge(order_details, on='order_id', how='inner')
print(f'df1 {df1.shape}')
df2 = df1.merge(products, on='product_id', how='inner')
print(f'df2 {df2.shape}')
df3 = df2.merge(categories, on='category_id', how='inner')
print(f'df3 {df3.shape}')
df4 = df3.merge(customers, on='customer_id', how='inner')
print(f'df4 {df4.shape}')
df5 = df4.merge(employees, on='employee_id', how='inner')
print(f'df5 {df5.shape}')
# df6 = df5.merge(employee_territories, on='employee_id', how='inner')
# print(f'df6 {df6.shape}')
# df7 = df5.merge(territories, on='territory_id', how='inner')
# print(f'df7 {df7.shape}')
# df8 = df7.merge(region, on='region_id', how='inner')
# print(f'df8 {df8.shape}')
df9 = df5.merge(suppliers, on='supplier_id', how='inner')
print(f'df9 {df9.shape}')
```
